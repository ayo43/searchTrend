```python
import pandas as pd
from pytrends.request import TrendReq
import matplotlib.pyplot as plt
trends = TrendReq()
```

```python
search ="Vaccine"
trends.build_payload(kw_list=[search])
data = trends.interest_by_region()
data = data.sort_values(by=search, ascending=False)
data = data.head(20)
print(data)
```

```python
data.reset_index().plot(x="geoName", y=search, figsize=(20,15), kind="bar")
plt.style.use('fivethirtyeight')
plt.show()
```

```python
data = TrendReq(hl='en-US', tz=360)
data.build_payload(kw_list=[search])
data = data.interest_over_time()
fig, ax = plt.subplots(figsize=(20, 15))
data[search].plot()
plt.style.use('fivethirtyeight')
plt.title('Total Google Searches for ' + ' ' + search, fontweight='bold')
plt.xlabel('Year')
plt.ylabel('Total Count')
plt.show()
```
